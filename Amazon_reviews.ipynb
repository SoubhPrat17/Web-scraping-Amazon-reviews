```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
website = requests.get(url)
soup = BeautifulSoup(website.content,'html.parser')
data = soup.findAll('div',attrs={'class':'a-section review aok-relative'})

names=[]
ratings=[]
titles_of_reviews=[]
written_reviews=[]

data    
```

```python
for i in data:
    name = i.find("span",class_="a-profile-name").text.replace('\n','')
    print(name)
    names.append(name)

    rating = i.find("span",class_='a-icon-alt').text.replace('\n','')
    print(rating)
    ratings.append(rating)

    title_of_review =i.find("a",class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold').text.replace('\n','')
    print(title_of_review)
    titles_of_reviews.append(title_of_review)

    written_review = i.find("span",class_='a-size-base review-text review-text-content').text.replace('\n','')
    print(written_review)
    written_reviews.append(written_review)  
```

```python
data_frame = pd.DataFrame({"Name":names,"Rating":ratings,"Review Title":titles_of_reviews,"Written reviews":written_reviews})
data_frame
```

```python

data_frame.to_csv('Amazon_reviews.csv')
```
